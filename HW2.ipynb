```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('/content/HW2_linear_data 2.csv')
X1 = data['X'].values
Y1 = data['Y'].values
# Initialize parameters (m and c)
m = 0
c = 0

# Hyperparameters
learning_rate = 0.0001
epochs = 1000
def mse_lossfunc(Y1, Y_pred):
    return ((Y1 - Y_pred) ** 2).mean()

# store the loss values for plotting
loss_history = []

# Training the model using gradient descent
for epoch in range(epochs):
    # Calculating predictions
    Y_pred = m * X1 + c

    # Calculate the gradient  with values to m and c
    dm = (-2 / len(X1)) * np.sum(X1* (Y1 - Y_pred))
    dc = (-2 / len(X1)) * np.sum(Y1- Y_pred)

    # Update  the parameters using gradient descent
    m -= learning_rate * dm
    c -= learning_rate * dc

    # store the current loss
    loss = mse_lossfunc(Y1, Y_pred)
    loss_history.append(loss)

# Plot the data points and the linear regression line
plt.scatter(X1, Y1, label='Data Points')
plt.plot(X1, m * X1 + c, color='red', label='Linear Regression Line')
plt.xlabel('X1')
plt.ylabel('Y1')
plt.title('Linear Regression using Gradient Descent')
plt.legend()
# Show the plot
plt.show()
# Print the  mse loss value
final_loss = loss_history[-1]
```
